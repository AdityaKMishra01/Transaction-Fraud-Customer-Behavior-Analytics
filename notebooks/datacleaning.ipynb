# 6. Data Type Conversion

Convert columns to appropriate data types, particularly the timestamp column, to enable accurate time-based analysis and feature engineering.

# 7. Feature Engineering

Create new analytical features that improve transaction analysis and fraud detection while maintaining the interpretability of the dataset.

## 7.1 Time-Based Features

Extract useful temporal information from the transaction timestamp, including month, day, and time period, to support trend analysis and fraud pattern identification.

```python
hours = [0, 5, 12, 17, 21, 24]
labels = ['Night', 'Morning', 'Afternoon', 'Evening', 'Night']

transactions['Time_period'] = pd.cut(transactions['timestamp'].dt.hour,bins=hours,labels=labels,ordered=False)
```

```python
transactions['amount (INR)'].min()
transactions['amount (INR)'].max()
```

```python
print(transactions['amount (INR)'].describe())
```

## 7.2 Transaction Amount Features

Categorize transaction amounts into meaningful groups and identify high-value transactions for further business analysis.

```python
conditions = [
    (transactions['amount (INR)'] <= 600),                      
    (transactions['amount (INR)'] > 600) & (transactions['amount (INR)'] <= 2000),
    (transactions['amount (INR)'] > 2000)                            
]

labels = ['Low', 'Medium', 'High']

transactions['amount_category'] = np.select(conditions, labels,default='Unknown')
```

```python
transactions['Same_Bank_Transaction'] = np.where(transactions['sender_bank'] == transactions['receiver_bank'],'Yes','No')
```
