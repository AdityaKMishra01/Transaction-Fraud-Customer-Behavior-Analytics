# Transaction Fraud & Customer Behavior Analytics

## Business Problem

Financial institutions process large volumes of digital payment transactions every day. As transaction volume grows, monitoring fraudulent activities and understanding customer transaction behavior become increasingly important.

Without proper analysis, organizations may struggle to identify high-risk transaction patterns, understand customer usage trends, prioritize fraud monitoring efforts, and make data-driven operational decisions.

This project aims to analyze transaction data to uncover fraud patterns, customer behavior, banking trends, and transaction insights that support business decision-making.

---

## Business Objectives

The primary objectives of this project are:

- Analyze transaction behavior across multiple business dimensions.
- Identify fraud distribution across transaction types, merchant categories, banks, devices, and locations.
- Understand customer demographics and transaction behavior.
- Evaluate banking transaction trends.
- Build interactive dashboards to support business decision-making.
- Generate actionable insights using Python, SQL, and Power BI.

---

## Project Scope

This project focuses on descriptive and diagnostic analytics using a 2024 transaction dataset.

The project includes:

- Data Understanding
- Data Cleaning
- Feature Engineering
- Exploratory Data Analysis (EDA)
- SQL Business Analysis
- Interactive Power BI Dashboard

This project does not include predictive machine learning models or real-time fraud detection.


---

## Dataset Overview

Dataset Name: Transactions Dataset (2024)

Source: Kaggle

Total Records: 250,000

Total Features: 17

The dataset contains information related to transaction details, customer demographics, banking information, merchant categories, transaction status, transaction amount, device usage, network type, timestamps, and fraud labels.

## Business Questions

### Transaction Analysis

- How many transactions were processed?
- What is the total transaction value?
- Which transaction types are most common?
- Which merchant categories process the highest transaction volume?
- What are the peak transaction hours?

### Fraud Analysis

- Which transaction types experience the highest fraud volume?
- Which merchant categories are associated with more fraud?
- Which banks record higher fraud activity?
- How does fraud vary by device type?
- How does fraud vary across network types?
- Which states report higher fraud activity?
- During which hours does fraud peak?

### Customer & Banking Analysis

- Which age groups are the most active users?
- How are transactions distributed across states?
- Are customers more likely to perform same-bank or cross-bank transactions?
- How does transaction activity differ between weekdays and weekends?
